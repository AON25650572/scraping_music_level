```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
url = "https://appmedia.jp/bang_dream/696454"
```

```python
# Requestsを利用してWebページを取得する
r = requests.get(url)
```

```python
# BeautifulSoupを利用してWebページを解析する
soup = BeautifulSoup(r.text, "html.parser")
```

```python
# テーブル形式のデータを取得（単体）
tbody = soup.find_all("tbody")[0]
```

```python
def tbody2pd(tbody):
    # tbody:テーブル形式のデータを取得（単体）
    
    # 行で分割
    trs = tbody.find_all("tr")

    # 1行目は基本的にheaderのはず
    header = []
    for head in trs[0].find_all("th"):
        if head.get("colspan"):
            for i in range(int(head.get("colspan"))):
                header.append(head.string)
        else:
            header.append(head.string)

    if len(header)==0:
        header = [i for i in range(len(trs[0].find_all("td")))]

    # Web表形式をPandasに変換
    df = pd.DataFrame(columns=header)

    for num,tr in enumerate(trs):
        inserts = []
        for td in tr.find_all("td"):
            inserts.append(td)
        if inserts:
            df.loc[num] = inserts

    return df
```

```python
# テーブルの中で["楽曲名", "追加方法"]のヘッダを持つ表を集めてくる
# 楽曲名の列から楽曲名とURLを抽出
url_dict = {}

for tbody in soup.find_all("tbody"):
    try:
        df = tbody2pd(tbody)
    
    except:
        # 表中にセル結合的なものには対応していない
        print("予期せぬ形式がありました")
        continue
    
    if list(df.columns)==["楽曲名", "追加方法"]:
        for minfos in df["楽曲名"]:
            minfo = minfos.find_all()[1]
            url_dict[minfo.string] = minfo.get("href")
```

```python
url_dict
```

```python
level_and_notes = []

for music_url in tqdm(url_dict.values()):
    
    # Requestsを利用してWebページを取得する
    r = requests.get(music_url)
    # BeautifulSoupを利用してWebページを解析する
    soup = BeautifulSoup(r.text, "html.parser")

    for tbody in soup.find_all("tbody"):
        try:
            df = tbody2pd(tbody)

        except:
            print("予期せぬ形式がありました")
            continue
            
        if list(df.columns)==["楽曲時間", "実装日"]:
            minits,seconds = df["楽曲時間"][1].string.split(":")
            total_seconds = int(minits) * 60 + int(seconds)

        if list(df.columns)==["譜面難易度", "楽曲Lv", "全ノーツ数"]:
            break

    for i,j in zip(df["楽曲Lv"],df["全ノーツ数"]):
        level_and_notes.append([int(i.string),int(j.string),total_seconds])
```

```python
with open("level_and_notes.csv","w") as f:
    for i,j,k in level_and_notes:
        f.write(str(i) + "," + str(j) + "," + str(k) + "\n")
```
